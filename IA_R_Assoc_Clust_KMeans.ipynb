```
%%R
library(cluster)
```

```
%%R
salario_conta <- read.csv('https://raw.githubusercontent.com/lcbjrrr/data/main/salario_conta%20.csv')
print(salario_conta)
plot(salario_conta$salario ~ salario_conta$saldo)
```

```
%%R
k <- kmeans(x = salario_conta,centers = 5)
print(k$cluster)
clusplot(x = salario_conta, clus = k$cluster, color = T, shade = T, labels = 5, lines = 0,main = "KMEANS")
```
