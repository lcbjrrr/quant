## Equação diferencial ordinária (EDO) numérica aplicada a equação de Solow - CES
Seja a equação do acúmulo de capital per capita 
\begin{equation}
 \dot{\kappa}(t) = sAf(\kappa) - (\delta + n) \kappa(t). \tag{Eq. do acúmulo de capital per capita}
\end{equation} 


## CES - Elasticidade de substituição constante 
Seja a função CES, $$Q(L, K)=(\alpha L^\rho+ (1-\alpha)K^\rho)^\frac{1}{\rho},$$ dividindo pela variável $L$, $$\frac{Q(L, K)}{L}=\frac{(\alpha L^\rho+ (1-\alpha)K^\rho)^\frac{1}{\rho}}{L}.$$ 

Definindo,
\begin{equation*}
\begin{aligned}
f(\kappa)&=&\left(\alpha+ (1-\alpha)\left(\frac{K}{L}\right)^\rho\right)^\frac{1}{\rho},\\
f(\kappa)&=&\left(\alpha+ (1-\alpha)\kappa^\rho\right)^\frac{1}{\rho}.
\end{aligned}
\end{equation*}

Substituindo $f(\kappa)$ na equação de acumulo de capital per capita,
\begin{equation}
 \dot{\kappa}(t) = sAf(\kappa) + (\delta + n) \kappa(t). \tag{Eq. do acúmulo de capital per capita}
\end{equation} 
teremos,
\begin{equation}
 \dot{\kappa}(t) = sA\left(\alpha+ (1-\alpha)\kappa^\rho\right)^\frac{1}{\rho} - (\delta + n) \kappa(t). \tag{Eq. de Solow-Swan CES}
\end{equation} 

## Solução numérica da equação de Solow - CES

```
def fun_solow_CES(t, y):
    s = 0.3
    A = 1
    alpha = 0.5
    delta = 0.1
    n = 0.015
    rho = 0.6
    fun_solow_CES = s * A * (alpha + (1 - alpha) * y**rho)**(1/rho) - (delta + n) * y;
    return fun_solow_CES   
#
import numpy as np
from numpy import exp
import matplotlib.pyplot as plt
a = 0
b = 200
n = 2048
h = (b - a) / n
t      = np.linspace(0, 0, n + 1)
y      = np.linspace(0, 0, n + 1)
sol_ex = np.linspace(0, 0, n + 1)
erro   = np.linspace(0, 0, n + 1)
for i in range(0, n + 1):
	t[i] = a + i * h
y[0] = 1.0
print(y[0])
for i in range(1, n + 1):
	y[i] = y[i - 1] + h * fun_solow_CES(t[i - 1], y[i - 1])
# Plot
plt.plot(t, y)
plt.title('EDO - Solow-CES')
plt.xlabel('t')
plt.ylabel('$\kappa(t)$')
plt.grid(alpha=.3, linestyle='--')
plt.show()
```

## Gráfico $\dot{\kappa}\times \kappa$


```
dot_y = np.linspace(0, 0, n + 1)
#for i in range(0, n + 1):
#	t[i] = a + i * h
for i in range(0, n + 1):
  dot_y[i] = fun_solow_CES(t[i], y[i])
# Plot
plt.plot(y, dot_y)
plt.title('EDO- Solow-CES')
plt.xlabel('$\kappa$')
plt.ylabel('$\dot{\kappa}(t)$')
plt.grid(alpha=.3, linestyle='--')
plt.show()
```
