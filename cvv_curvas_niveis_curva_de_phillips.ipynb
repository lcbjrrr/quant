# Função perda do Banco central
Seja a função perda do banco central,
\begin{equation}
\begin{aligned}
L(y_t, \pi_t)&=&(y_t-y_e)^2 + \beta(\pi_t-\pi^T)^2, \\
\end{aligned}\tag{Função perda do banco central}
\end{equation}
sujeita a
\begin{equation}
\begin{aligned}
\pi_{t}&=&\pi_{t-1}+\alpha\left(y_t-y_e\right),
\end{aligned}\tag{Curva de Phillips-PC}
\end{equation}
onde $y_t$ e $\pi_t$ são respectivamente o produto e a inflação no tempo corrente, e $y_e$ é o produto esperado, e o $\pi^T$ é a inflação alvo. Os parâmetros são $\alpha$ e $\beta>0$.

O problema será resolvido com técnicas de cálculo I, com esse objetivo, vamos substituir a equação chamada de Curva de Phillips na função perda do banco central. Obteremos
\begin{equation}
\begin{aligned}
L(y_t)&=&(y_t-y_e)^2 + \beta(\pi_{t-1}+\alpha\left(y_t-y_e\right)-\pi^T)^2.\\
\end{aligned}\tag{Curva paramétrica PC}
\end{equation}

Encontramos o ponto crítico na Curva parametrica para PC
\begin{equation}
\begin{aligned}
\frac{dL(y_t)}{d y_t}&=&(y_t-y_e) + \alpha\beta(\pi^T+\alpha\left(y_t-y_e\right)-\pi^T)=0,\\
\end{aligned}
\end{equation}
reescrevendo a curva de Phillips, teremos
\begin{equation}
\begin{aligned}
(y_t-y_e) + \alpha\beta(\pi_{t}-\pi^T)&=&0, \\
\end{aligned}
\end{equation}
e verifivando a segunda derivada de $L(y_t)$ em relação a $y_t$,
\begin{equation}
\begin{aligned}
\frac{d^2L(y_t)}{d y_t^2}&=1 + \alpha^2\beta>0,\\
\end{aligned}
\end{equation}

portanto a regra monetária
\begin{equation}
\begin{aligned}
(y_t-y_e) &=&-\alpha\beta(\pi_{t}-\pi^T) \\
\end{aligned}\tag{Regra monetária-MR}
\end{equation}
é deduzida de um problema de otimização
\begin{equation*}
\begin{aligned}
& \underset{y_t,\pi_t}{\text{maximize}}
& L(y_t, \pi_t) & \\
& \text{sujeito a}
& \pi_{t}-\pi_{t-1}-\alpha\left(y_t-y_e\right) & =0.
\end{aligned}
\end{equation*}

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as pyplot
import numpy as numpy
import numpy as np
# a = produto estimado
# b = inflação alvo
a = 2
b = 2
epsilon = 0.00001
x = np.outer(np.linspace(0, 4, 32), np.ones(32))
y = x.copy().T
z = (x - a)**2 + (y - b)**2  # Função perda
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(x, y, z,
					rstride = 1,
					cstride = 1,
					alpha = 0.4,
					color= 'lightgreen')
#  Gráfico reta 1
x1 = np.linspace(0, 4, 100)
y1 = x1
ax.plot(x1, y1, zs=0, zdir='z', color='blue')
#  Gráfico reta 2
x1 = np.linspace(0, 3, 100)
y1 = x1 + np.sqrt(2)
ax.plot(x1, y1, zs=0, zdir='z', color='blue')
#  Plot reta 3
x1 = np.linspace(0, 4, 100)
y1 = -x1 + 4
ax.plot(x1, y1, zs=0, zdir='z', color='orange')
z2 = 1
x2 = np.linspace(a - np.sqrt(z2), a + np.sqrt(z2), 100)
y2 =  np.sqrt(z2 - (x2 - a)**2) + b
y3 = -np.sqrt(z2 - (x2 - a)**2) + b
#ax.plot(x2, y2, zs=0, zdir='z', color='blue')
#ax.plot(x2, y3, zs=0, zdir='z', color='blue')
#
ax.plot(x2, y2, zs=1, zdir='z', color='green')
ax.plot(x2, y3, zs=1, zdir='z', color='green')
# plot curva de nível 2
z2 = 2
x2 = np.linspace(a - np.sqrt(z2) + epsilon, a + np.sqrt(z2), 100)
y2 =  np.sqrt(z2 - (x2 - a)**2) + b
y3 = -np.sqrt(z2 - (x2 - a)**2) + b
#ax.plot(x2, y2, zs=0, zdir='z', color='blue')
#ax.plot(x2, y3, zs=0, zdir='z', color='blue')
#
ax.plot(x2, y2, zs=2, zdir='z', color='green')
ax.plot(x2, y3, zs=2, zdir='z', color='green')
# plot curva de nível 3
z2 = 3
x2 = np.linspace(a - np.sqrt(z2) + epsilon, a + np.sqrt(z2), 100)
y2 =  np.sqrt(z2 - (x2 - a)**2) + b
y3 = -np.sqrt(z2 - (x2 - a)**2) + b
#ax.plot(x2, y2, zs=0, zdir='z', color='blue')
#ax.plot(x2, y3, zs=0, zdir='z', color='blue')
#
ax.plot(x2, y2, zs=3, zdir='z', color='green')
ax.plot(x2, y3, zs=3, zdir='z', color='green')
#
theta = np.linspace(0.6, 2.0, 100)
xc = theta
yc = xc + np.sqrt(2)
zc = (xc - a)**2 + (yc - b)**2
ax.plot(xc, yc, zc, color='red')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.xlabel('y   ')
ax.set_ylabel('$\pi$')
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.set_zlabel('$L(\pi,y)$')
ax.set_zlim(np.min(z), np.max(z))
ax.set_title('Banco Central - Função Perda')
plt.show()
```
