# Altair Lib

```python
import altair as alt
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/grades%20-%20okk.csv')
df.head(2)
```

# Bar chart

```python
alt.Chart(df).mark_bar().encode(x='Student',y='Grade')
```

```python
grade_per_course = df.groupby('Course').mean()['Grade']
grade_per_course
```

```python
grade_per_course = grade_per_course.reset_index()
grade_per_course
```

```python
alt.Chart(grade_per_course).mark_bar().encode(x='Course',y='Grade')
```

```python
alt.Chart(df).mark_bar().encode(x='Course',y='mean(Grade)')
```

```python
alt.Chart(df).mark_bar().encode(x='Course',y='Grade')
```

# Pie Chart

![](https://pbs.twimg.com/media/Fw7gCt_XgAEcUk0?format=png&name=small)

```python
alt.Chart(df).mark_arc().encode(theta='count(Student)', color='Course')
```

# Scatterplot

```python
grades = pd.read_csv('https://raw.githubusercontent.com/lcbjuk/ML/master/dados/notas%20-%20ok.csv')
grades.head()
```

```python
alt.Chart(grades).mark_point().encode(
    x='AP2',
    y='AP3'
)
```

```python
alt.Chart(grades).mark_point().encode(
    x='AP2',
    y='AP3'
).interactive()
```

# Heatmap / Correlation

```python
cor = grades.corr()
cor=cor.reset_index()
cor
```

```python
corr = cor.melt(id_vars=['index'])
corr.head(3)
```

```python
alt.Chart(corr).mark_rect().encode(
    x='index',
    y='variable',
    color='value'    
)
```
