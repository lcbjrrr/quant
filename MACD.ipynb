```
import pandas as pd

#petr4 = pd.read_csv('/content/drive/My Drive/IBMEC/QUANT/petr4.sa - ok.csv')
petr4 = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/petr4.sa%20-%20ok.csv')
petr4['Close']
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot( petr4['Date'] , petr4['Close'])
plt.show()
```

```
longa = petr4['Close'].ewm(span=26 , adjust= False).mean()
curta = petr4['Close'].ewm(span=12 , adjust= False).mean()
plt.figure(figsize=(20,10))
plt.plot(petr4['Date'] , longa , label='Longa' , color='red')
plt.plot(petr4['Date'] , curta , label='Curta', color='blue')
plt.legend(loc='upper left')
plt.show()
```

```
macd = curta-longa
sinal = macd.ewm(span=9 , adjust= False).mean()
plt.figure(figsize=(20,10))
plt.plot(petr4['Date'] , macd , label='MACD' , color='red')
plt.plot(petr4['Date'] , sinal , label='Sinal', color='blue')
plt.legend(loc='upper left')
plt.show()
```
