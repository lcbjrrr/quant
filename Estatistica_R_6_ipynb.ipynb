**Regressao Linear**



```
%%R
nts <- read.csv('https://raw.githubusercontent.com/lcbjuk/ML/master/dados/notas%20-%20ok.csv')
head(nts)

```

```
%%R
reglin <- lm(nts$AP3 ~ nts$AP1 + nts$AP2)
reglin
```

```
%%R
prever_ap3 <- function(ap1, ap2){
  ap3 <- ap1 * -0.366 + ap2 * 1.383 - 0.825
  return(ap3)
}

print(prever_ap3(8,8))
print(prever_ap3(2,4))
print(prever_ap3(3.5,8))
```

```
%%R
plot(nts$AP2 , nts$AP3)
abline(lm(nts$AP3~nts$AP2))
```

```
%%R
cor(nts)
```

```
%%R
cor.test(nts$AP3,nts$AP2)
```
