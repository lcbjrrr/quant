# Equação de Cobb-Douglas
\begin{equation}
\begin{aligned}
P(L, K) &= \beta L^\alpha K^{1-\alpha}
\end{aligned}\tag{Eq. de Cobb-Douglas}
\end{equation}

## Parâmetros para o programa
- $\beta=1$
- $\alpha=0.75$

## Gráfico da função Cobb-Douglas

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Preparando dados.
X = np.arange(0, 300, 1)
Y = np.arange(0, 300, 1)
X, Y = np.meshgrid(X, Y)

alpha = 0.75
beta = 1 - alpha
b = 1.0

Z = b * X**(alpha) * Y**(beta)

# Plotando.
surf = ax.plot_surface(X, Y, Z, cmap='RdGy',
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(0, 300)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
# Barra de cores.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('Gráfico da função Cobb-Douglas')
plt.xlabel('L')
plt.ylabel('K')
plt.show()
```

## Curvas de Níveis
Seja a função de Cobb-Douglas
\begin{equation}
\begin{aligned}
P(L, K) &= \beta L^\alpha K^{1-\alpha},
\end{aligned}
\end{equation}
para encontrar as curvas de níveis é necessário fixar níveis de produção.

Para o nível de produção $P_n$
\begin{equation}
\begin{aligned}
P_n &= \beta L^\alpha K^{1-\alpha},\\
K^{1-\alpha} &=\frac{P_n} {\beta L^\alpha},
\end{aligned}
\end{equation}
portanto, para cada nível de produção $P_n$ dado teremos a curva
\begin{equation}
\begin{aligned}
K &=\left[\frac{P_n} {\beta L^\alpha}\right]^{\frac{1}{1-\alpha}}.\end{aligned}
\end{equation}

O programa mostra as curvas de níveis (isoquantas) para a função de Cobb-Douglas.

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
def f(x, y):
    alpha = 0.75
    beta = 1 - alpha
    b = 1.0
    return b * x**(alpha) * y**(beta)

x = np.linspace(0, 300, 50)
y = np.linspace(0, 300, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
contours = plt.contour(X, Y, Z, 10, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 300, 0, 300], origin='lower',
           cmap='RdGy')
plt.colorbar();
plt.title('Curva de níveis- Cobb-Douglas')
plt.xlabel('L')
plt.ylabel('K')
plt.show()
```
