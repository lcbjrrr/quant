# Comparacao

## Py

```python
import pandas as pd 
cases = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/covid19%20-%20cases-melt.csv')
cases.head(1)
```

```python
max_casos_por_pais = cases.groupby('Country').max()['cases']
max_casos_por_pais
```

```python
max_casos_por_pais.plot(kind='bar')
```

## R

```python
%load_ext rpy2.ipython
```

```python
%%R
casos<-read.csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/covid19%20-%20cases-melt.csv')
head(casos,1)
```

```python
%%R
max_casos_por_pais <- tapply(casos$cases, casos$Country, max)
max_casos_por_pais
```

```python
%%R
barplot(max_casos_por_pais)
```
